```
tuned_dt = tune_model(et)
```

```
plot_model(tuned_dt, plot = 'confusion_matrix')
# Decision Trees did a pretty great job with only 4 wrong predictions.
```

```
predict_model(tuned_dt);
```

```
final_et = finalize_model(tuned_dt)
```

```
preds = predict_model(final_et, data=test.drop(columns=['fname','label'],axis=1))
```

# STACKED MODEL - ['ExtraTreesClassifier', 'RandomForestClassifier' and 'XGBoost']

```
# create individual models for stacking
et = create_model('et')
rf = create_model('rf')
xgboost = create_model('xgboost')
# stacking models
stacker = stack_models(estimator_list = [et,rf], meta_model = xgboost)
```

```
tuned_nb = tune_model(stacker)
plot_model(tuned_nb, plot = 'confusion_matrix')
evaluate_model(tuned_nb)
predict_model(tuned_nb);

# The model seems to have done a great job and we will use this for submission.
```

```
final_nb = finalize_model(tuned_nb)
preds = predict_model(final_nb, data=test.drop(columns=['fname','label'],axis=1))
final_preds = preds['Label']

sub = pd.DataFrame()
sub['Filename'] = test.fname
sub['Class'] = final_preds

sub = sub.sort_values(by='Filename')
sub.to_csv('hack_nitp_pycaret_stacker_v3.csv',index=False)
```
